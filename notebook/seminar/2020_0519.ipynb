```
import numpy as np
import wave
import struct
import matplotlib.pyplot as plt

def showDetails(wr):
    # waveファイルが持つ性質を取得
    ch = wr.getnchannels()
    width = wr.getsampwidth()
    fr = wr.getframerate()
    fn = wr.getnframes()

    print("Channel: ", ch)
    print("Sample width: ", width)
    print("Frame Rate: ", fr)
    print("Frame num: ", fn)
    print("Params: ", wr.getparams())
    print("Total time: ", 1.0 * fn / fr)

def showGraph(wr, start = 0, end = -1, frame_rate = 16000, title="title"):
    
    if end == -1:
        end = wr.getnframes()

    # waveの実データを取得し、数値化
    wr.rewind()
    buf = wr.readframes(wr.getnframes())
    data = np.frombuffer(buf, dtype = "int16")

    # 音声データ全体のグラフ
    x = [i / frame_rate for i in range(start, end)]
    y = data[start: end]

    plt.plot(x, y)
    plt.title(title)
    plt.show()
```

```
# ファイルを読み出し
wr = wave.open('a.wav', 'r')
showDetails(wr)
showGraph(wr,title = "Ah(All)")
showGraph(wr, 800, 1600, title = "Ah(0.05 ~ 0.1)")
wr.close()
```

```
# ファイルを読み出し
wr = wave.open('i.wav', 'r')
showDetails(wr)
showGraph(wr,title = "Eeh(All)")
showGraph(wr, 800, 1600, title = "Eeh(0.05 ~ 0.1)")
wr.close()
```
