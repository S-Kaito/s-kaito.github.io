```
import numpy as np
import wave
import struct
import matplotlib.pyplot as plt

START = 800
END = 1600

# ファイルを読み出し
wavf = 'a.wav'
wr = wave.open(wavf, 'r')

# waveファイルが持つ性質を取得
ch = wr.getnchannels()
width = wr.getsampwidth()
fr = wr.getframerate()
fn = wr.getnframes()

print("Channel: ", ch)
print("Sample width: ", width)
print("Frame Rate: ", fr)
print("Frame num: ", fn)
print("Params: ", wr.getparams())
print("Total time: ", 1.0 * fn / fr)

# waveの実データを取得し、数値化
buf = wr.readframes(wr.getnframes())
data = np.frombuffer(buf, dtype = "int16")
wr.close()

# 音声データ全体のグラフ
X = [i / 16000 for i in range(0, fn)]
Y = data[: ]

plt.plot(X, Y)
plt.title("voice: Ah~~(All)")
plt.show()

# 0.05 ~ 0.1 秒のデータのグラフ
x = [i / 16000 for i in range(START, END)]
y = data[START: END]

plt.plot(x, y)
plt.title("voice: Ah~~(0.05 ~ 0.1 sec)")
plt.show()
```
