```
import numpy as np
import wave
import struct
import matplotlib.pyplot as plt

def showDetails(wr):
    """
    Show wave files details

    Parameter
    --------
    wr: wave object
        wave object
    """
    # waveファイルが持つ性質を取得
    ch = wr.getnchannels()
    width = wr.getsampwidth()
    fr = wr.getframerate()
    fn = wr.getnframes()

    print("Channel: ", ch)
    print("Sample width: ", width)
    print("Frame Rate: ", fr)
    print("Frame num: ", fn)
    print("Params: ", wr.getparams())
    print("Total time: ", 1.0 * fn / fr)


def showGraph(wr, start = 0, end = -1, title="title"):
    """
    Show graph

    Parameter
    ----------
    wr: wave object
        wave object
    start: int
        time(sec)
    end: int
        time(sec)
    title: string
        graph title
    """
    fr = wr.getframerate()
    fn = wr.getnframes()

    if end == -1:
        end = 1.0 * fn / fr

    start_f = int(fr * start)
    end_f = int(fr * end)

    # waveの実データを取得し、数値化
    wr.rewind()
    buf = wr.readframes(wr.getnframes())
    data = np.frombuffer(buf, dtype = "int16")

    # 音声データ全体のグラフ
    x = [i / fr for i in range(start_f, end_f)]
    y = data[start_f: end_f]

    plt.plot(x, y)
    plt.title(title)
    plt.show()
```

```
# ファイルを読み出し
wr = wave.open('a.wav', 'r')
showDetails(wr)
showGraph(wr,title = "Ah(All)")
showGraph(wr, 0.05, 0.1, title = "Ah(0.05 ~ 0.1)")
showGraph(wr, 0.05, 0.07, title = "Ah(0.05 ~ 0.07)")
wr.close()
```

```
# ファイルを読み出し
wr = wave.open('i.wav', 'r')
showDetails(wr)
showGraph(wr,title = "Eeh(All)")
showGraph(wr, 0.05, 0.1,title = "Eeh(0.05 ~ 0.1)")
showGraph(wr, 0.05, 0.07,title = "Eeh(0.05 ~ 0.07)")
wr.close()
```
