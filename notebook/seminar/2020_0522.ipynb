```
import sys
import struct
import numpy as np
import codecs
from PIL import Image
import numpy as np

def read_picture(path):
    """
    Read picture from file

    Parameter
    ---------
    path: string
        file path
    
    return
    ------
    size: (width: int, height: int)
        picture size
    max_n: int
        max value
    data: Array<Array<[r: int, g: int, b: int]>>
        picture data
    """
    
    # 画像を開く(Header情報用)
    infile = codecs.open(path, 'r', "utf-8", "ignore")
    # 画像を開く(画像データ用)
    img = np.array(Image.open(path))

    infile.readline()
    infile.readline()
    width, height = list(map(int, infile.readline().split()))
    max_n = int(infile.readline())

    # データを(width * height, 3)に変形
    data =[]
    for im in img:
        for i in im:
            r = i[0]
            g = i[1]
            b = i[2]
            data.append([r, g, b])

    infile.close()
    return (width, height), max_n, data

def write_picture(path, size, n, data):
    """
    Write picture to file
    Parameter
    ---------
    path: string
        file path
    size: (width: int, height: int)
        picture size
    n: int
        max value
    data: Array<Array<[r: int, g: int, b: int]>>
        picture data
    """

    # 画像を書き出す
    # その際に1次元でデータを扱っているので2次元に変換する
    pilImg = Image.fromarray(np.uint8(np.array(data).reshape((size[0], size[1]))))
    pilImg.save(path)

def extract_red(data):
    # [R, G, B]のRのみを抽出する
    res = [d[0] for d in data]
    return res

def extract_green(data):
    # [R, G, B]のGのみを抽出する
    res = [d[1] for d in data]
    return res

def extract_blue(data):
    # [R, G, B]のBのみを抽出する
    res = [d[2] for d in data]
    return res

def extract_average(data):
    # [R, G, B]の平均で生成する
    res = [np.average(d) for d in data]
    return res

def parse_gray_scale(data):
    # NTSC形式用の配列
    value = np.array([
        [0.299, 0.587, 0.114],
        [0.596, -0.274, -0.322],
        [0.212, -0.523, 0.311]
    ])

    #行列の掛け算を行い、そのうちのYのみで配列を生成する
    res = [np.dot(value, np.array(d)).tolist()[0] for d in data]

    return res
```
